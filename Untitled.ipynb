```python
def get_sample_object(ids=20312, 
                      ANNOTATIONS = ANNOTATIONS_DIR,
                      INSTANCE_ANNOTATIONS = INSTANCE_ANNOTATIONS_DIR,
                      DATA_S2 = DATA_S2_DIR):
    """Get object images and realated targets"""
    s2        = np.load(DATA_S2/f'S2_{ids}.npy')
    
    parcel    = np.load(ANNOTATIONS/f'ParcelIDs_{ids}.npy')
    target    = np.load(ANNOTATIONS/f'TARGET_{ids}.npy')
    
    instances = np.load(INSTANCE_ANNOTATIONS/f'INSTANCES_{ids}.npy')
    heatmap   = np.load(INSTANCE_ANNOTATIONS/f'HEATMAP_{ids}.npy')
    zones     = np.load(INSTANCE_ANNOTATIONS/f'ZONES_{ids}.npy')
    return s2, parcel, target, instances, heatmap, zones
    
```

```python
def get_rgb(x, batch_index=0, t_show=1):
    """Utility function to get a displayable rgb image 
    from a Sentinel-2 time series.
    """
    im = x[t_show, [2,1,0]].cpu().numpy()
    mx = im.max(axis=(1,2))
    mi = im.min(axis=(1,2))   
    im = (im - mi[:,None,None])/(mx - mi)[:,None,None]
    im = im.swapaxes(0,2).swapaxes(0,1)
    im = np.clip(im, a_max=1, a_min=0)
    return im


def plot_dataset_object(s2=None, parcel=None, target=None, instances=None, heatmap=None, zones=None):
    fig, ax = plt.subplots(2, 3,figsize=(16, 8))
    
    if s2 is not None:
        s2 = get_rgb(torch.from_numpy(s2))
        ax[0, 0].imshow(s2)
        ax[0, 0].title.set_text('S2 IMAGE')
    
    if parcel is not None:
        ax[0, 1].imshow(parcel)
        ax[0, 1].title.set_text('Parcel IMAGE')
    
    if target is not None:
        target = np.transpose(target, (1,2,0))
        ax[0, 2].imshow(target, cmap = cmap, vmin=0,vmax=20)
        ax[0, 2].title.set_text('Target IMAGE')
    
    if instances is not None:
        ax[1, 0].imshow(instances, cmap='prism', alpha=.6)
        ax[1, 0].title.set_text('Instances IMAGE')
         
    if heatmap is not None:
        ax[1, 1].imshow(heatmap)
        ax[1, 1].title.set_text('Heatmap IMAGE')
        
    if zones is not None:
        ax[1, 2].imshow(zones)
        ax[1, 2].title.set_text('Zones IMAGE')
        
        
        
    
    
```

```python
def plot_random_ds_object(seed=None):
    if seed is not None:
        np.random.seed(seed)
        
    ids = np.random.choice(metadata.ID_PATCH.values)

    rand_s2, rand_parcel, rand_target, rand_inst, rand_hmap, rand_zones = get_sample_object(ids=ids)
    plot_dataset_object(s2=rand_s2, parcel=rand_parcel, target=rand_target,instances=rand_inst,heatmap=rand_hmap, zones=rand_zones)
```
